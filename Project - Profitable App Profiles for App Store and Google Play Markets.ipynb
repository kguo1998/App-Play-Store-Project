```python
def explore_data(dataset, start, end, rows_and_columns):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))

def open_data(filename):
    opened = open(filename)
    from csv import reader
    read_file = reader(opened)
    data = list(read_file)
    return data

def is_English(string):
    non_ascii = 0
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
        
    if non_ascii > 3:
        return False
    else:
        return True

app_store = open_data('AppleStore.csv')
play_store = open_data('googleplaystore.csv')

explore_data(app_store, 1, 4, rows_and_columns=True)
explore_data(play_store, 1, 4, rows_and_columns=True)

print('\nApp Store Column Names:', app_store[:1], '\n')
print('Play Store Column Names:', play_store[:1], '\n')

### removing wrong data

del play_store[10473]

for row in play_store:
    header_len = len(play_store[0])
    row_length = len(row)
    if row_length != header_len:
        print(row)
        print(play_store.index(row))
        
### removing duplicates
        
reviews_max = {}

for row in play_store:
    name = row[0]
    n_reviews = row[3]
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

android_clean = []
already_added = []

for row in play_store:
    name = row[0]
    n_reviews = row[3]
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)
        
###explore_data(android_clean, 0, 3, True)

### removing non-english apps

android_english = []
ios_english = []

for row in android_clean:
    name = row[0]
    if is_English(name):
        android_english.append(row)
    
for row in app_store:
    name = row[1]
    if is_English(name):
        ios_english.append(row)

### isolating the free apps
android_final = []
ios_final = []

for row in android_english:
    price = (row[7])
    if price == '0':
        android_final.append(row)
        
for row in ios_english:
    price = (row[4])
    if price == '0.0':
        ios_final.append(row)
        

###explore_data(android_final, 1, 4, True)
###explore_data(ios_final, 1, 4, True)

### Buildling frequency tables on genres

def freq_table(dataset, index):
    table = {}
    total = 0
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1

    table_percentages = {}
    for key in table:
        percentage = (table[key]/total)*100
        table_percentages[key] = percentage
        
    return table_percentages
    
### Displaying percentages in a descending order

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
    table_sorted = sorted(table_display, reverse = True)
    for row in table_sorted:
        print(row[1], ':', row[0])

display_table(ios_final, -5)
print('\n\n')
display_table(android_final, 1)

### Generating freq table for each unique genre
genres_ios = freq_table(ios_final, -5)

### Looping through frequency table to find average number of ratings
print('\n')
for genre in genres_ios:
    total = 0
    gen_len = 0
    
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:
            n_ratings = float(app[5])
            total += n_ratings
            gen_len += 1

### Compute and display average ratings for each genre
    avg_rating = total / gen_len
    print(genre, ':', avg_rating)

### Freq table for unique app genres
print('\n')
categories_android = freq_table(android_final, 1)

for category in categories_android:
    total = 0
    cat_len = 0
    
    for app in android_final:
        app_category = app[1]
        if app_category == category:
            n_installs = app[5]
            n_installs = n_installs.replace(',','')
            n_installs = n_installs.replace('+','')
            total += float(n_installs)
            cat_len += 1
            
    avg_installs = total / cat_len
    print(category, ':', avg_installs)
   
```
